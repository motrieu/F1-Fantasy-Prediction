Import Libraries

```python
import pandas
import requests
from bs4 import BeautifulSoup
```

Create Array for Driver and Calendar

```python
track_list = set()
track_request = requests.get("https://www.formula1.com/en/racing/2022.html")
track_soup = BeautifulSoup(track_request.text, 'html.parser')
track_cards = track_soup.find_all("div", "event-place")
for card in track_cards:
     track_list.add(card.text)

driver_list = set()
driver_request = requests.get("https://www.formula1.com/en/drivers.html")
driver_soup = BeautifulSoup(driver_request.text, "html.parser")
driver_cards = driver_soup.find_all("div", "listing-item--name")
for card in driver_cards:
    driver_list.add(card.text)

track_list = list(track_list)
driver_list = list(driver_list)
```

Create Dataframe for 2022 driver by track

```python
driver_track_frame = pandas.DataFrame(index=driver_list, columns=track_list)
```
